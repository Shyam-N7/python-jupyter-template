### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

import numpy as np
import matplotlib.pyplot as plt

# Validation functions
def validate_data(x, y):
    if len(x) != len(y):
        raise ValueError("X and Y must have the same length.")
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise ValueError("X and Y values must be numeric.")

def validate_bar_data(categories, values):
    if len(categories) != len(values):
        raise ValueError("Categories and values must have the same length.")
    if not all(isinstance(v, (int, float)) for v in values):
        raise ValueError("Values must be numeric.")

def validate_scatter_data(x, y):
    if len(x) != len(y):
        raise ValueError("X and Y must have the same length.")
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise ValueError("X and Y values must be numeric.")

def validate_pie_data(labels, sizes):
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes must be of equal length.")
    if not all(isinstance(i, (int, float)) and i > 0 for i in sizes):
        raise ValueError("Sizes must be positive numbers.")

try:
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))

    # Line Chart
    x = np.arange(0, 11, 1)
    y = x**2
    validate_data(x.tolist(), y.tolist())
    axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')
    axes[0, 0].set_title("Line Chart")

    # Bar Chart
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    validate_bar_data(categories, values)
    axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
    axes[0, 1].set_title("Bar Chart")

    # Scatter Plot
    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]
    validate_scatter_data(x_scatter, y_scatter)
    axes[1, 0].scatter(x_scatter, y_scatter, color='purple', marker='o', label='Data points')
    axes[1, 0].set_title("Scatter Plot")

    # Pie Chart
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    validate_pie_data(labels, sizes)
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, wedgeprops={'edgecolor': 'black'})
    axes[1, 1].set_title("Pie Chart")

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
